```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('oxygene.csv')

fig = px.scatter_3d(df, x='latitude', y='longitude', z='depth', color='oxygene', opacity=0.7, width=600, height=600, title="Concentration d'oxygène en fonction des coordonnées et de la prodindeur (en m)")
fig.update_layout(margin=dict(l=50, r=50, b=50, t=50))
fig.show()
```

```python
fig = px.scatter(df, x = 'oxygene', y = 'depth', color="oxygene", height=800, marginal_x="histogram", marginal_y="rug", trendline="ols",title='titre')
fig.show()
fig.write_html("scatter_oxygene.html")
```
